```python
### âœ… 1. í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ëª¨ë“ˆ ë¶ˆëŸ¬ì˜¤ê¸°
import os
from process_video import extract_frames
from predict_deepfake_model import process_all_frames
import json
from IPython.display import display, Image

root_folder = "."

video_path = os.path.join(root_folder, "video/video.mp4")  # ë¶„ì„í•  ì˜ìƒ ê²½ë¡œ
frames_folder = os.path.join(root_folder, "frames")               # í”„ë ˆì„ ì €ì¥ í´ë”
```

```python
### âœ… 2. ì˜ìƒ â†’ í”„ë ˆì„ ì¶”ì¶œ ì…€
def extract_video_frames(video_path, output_folder="frames", fps=1):
    print(f"ğŸ¬ ì˜ìƒì—ì„œ í”„ë ˆì„ ì¶”ì¶œ ì¤‘...")
    extract_frames(video_path, output_folder, fps)
    print(f"âœ… í”„ë ˆì„ ì¶”ì¶œ ì™„ë£Œ: {output_folder}")

# ì‚¬ìš© ì˜ˆì‹œ
extract_video_frames(video_path, frames_folder, fps=15)
```

```python
### âœ… 3. ë”¥í˜ì´í¬ ì˜ˆì¸¡ (GradCAM ì‚¬ìš© ì—¬ë¶€ í¬í•¨)
def run_deepfake_prediction(root_folder, output_file="results/deepfake_results.json", batch_size=16, use_gradcam=True):
    print(f"ğŸ¤– ë”¥í˜ì´í¬ ì˜ˆì¸¡ ì¤‘... (GradCAM: {use_gradcam})")
    process_all_frames(root_folder, output_file, batch_size, use_gradcam)
    print(f"âœ… ì˜ˆì¸¡ ì™„ë£Œ: {output_file}")

# ì‚¬ìš© ì˜ˆì‹œ
output_json = "results/deepfake_results.json"
run_deepfake_prediction(root_folder, output_json, batch_size=16, use_gradcam=True)
```

```python
### âœ… 4. ê²°ê³¼ í™•ì¸ ë° GradCAM ì‹œê°í™”
def display_prediction_results(output_json, show_gradcam=False):
    import json
    import os
    import matplotlib.pyplot as plt
    from PIL import Image

    with open(output_json, "r") as file:
        data = json.load(file)

    # í”„ë ˆì„ ì´ë¦„(í‚¤)ì„ ì •ë ¬í•˜ì—¬ ìˆœì„œëŒ€ë¡œ ì²˜ë¦¬
    frames = sorted(data.keys())
    probabilities = []
    frame_numbers = []
    gradcam_first = None
    gradcam_frame = None


    for frame in frames:
        result = data[frame]
        # ê²°ê³¼ê°€ floatì¸ ê²½ìš°ì™€ dictì¸ ê²½ìš°ë¥¼ ëª¨ë‘ ê³ ë ¤í•˜ì—¬ í™•ë¥  ì¶”ì¶œ
        prob = result if isinstance(result, float) else result.get("probability", 0.0)
        probabilities.append(prob * 100)  # ë°±ë¶„ìœ¨ë¡œ ë³€í™˜

        # íŒŒì¼ ì´ë¦„ì—ì„œ ìˆ«ì ì¶”ì¶œ (ì˜ˆ: "frame_0001.jpg" -> 1)
        try:
            base = os.path.splitext(frame)[0]  # "frame_0001"
            num_str = base.split('_')[-1]
            frame_number = int(num_str)
        except Exception:
            frame_number = frame  # íŒŒì‹± ì‹¤íŒ¨ ì‹œ ì›ë˜ ì´ë¦„ ì‚¬ìš©
        frame_numbers.append(frame_number)

        # ì²« ë²ˆì§¸ GradCAM ì´ë¯¸ì§€ ê²½ë¡œ í™•ë³´ (ì¡´ì¬í•  ê²½ìš°)
        if show_gradcam and gradcam_first is None:
            gradcam_path = result.get("gradcam_path")
            if gradcam_path and os.path.exists(gradcam_path):
                gradcam_first = gradcam_path
                gradcam_frame = frame_number  # â† ë³€ê²½: ì›ë˜ frameì—ì„œ frame_numberë¡œ ìˆ˜ì •

    # í™•ë¥  ê·¸ë˜í”„ í”Œë¡¯
    plt.figure(figsize=(10, 5))
    plt.plot(frame_numbers, probabilities, marker='o', linestyle='-', color='b')
    plt.xlabel("Frame Number")
    plt.ylabel("Deepfake Probability (%)")
    plt.title("Deepfake Prediction per Frame")
    plt.xticks(rotation=90)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # ì²« ë²ˆì§¸ GradCAM ì´ë¯¸ì§€ í‘œì‹œ
    if show_gradcam and gradcam_first is not None:
        img = Image.open(gradcam_first)
        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.title(f"GradCAM for Frame {gradcam_frame}")
        plt.axis('off')
        plt.show()

# ì‚¬ìš© ì˜ˆì‹œ
display_prediction_results(output_json, show_gradcam=True)

```
